```python
# Create date column extracting only the yyyy-mm-dd from the 'time' column
data['date'] = pd.to_datetime(data['time']).dt.date

# Create month column extracting only the yyyy-mm-dd from the 'time' column
data['month'] = pd.to_datetime(data['time']).dt.month

# Create hour column extracting on the hh from the 'time' column
data['hour'] = pd.to_datetime(data['time']).dt.hour
```

```python
# Convert magnitudes to a numpy array
magnitudes = np.array(data['mag'])

# Create a figure
fig = plt.figure()

# Create subplot
ax = fig.subplots()

# Create a histogram (bar plot)
ax.hist(magnitudes, bins=50, color='green', edgecolor='black')

# Add title and labels
ax.set_title('Distribution of Magnitudes')
ax.set_xlabel('Magnitude')
ax.set_ylabel('Frequency')


plt.tight_layout()
```

```python
# Count occurrences of each place
place_counts = data['place'].value_counts()

# Display the top 10 places with the most earthquakes
top_places = place_counts.head(10)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
ax.bar(top_places.index, top_places.values, color='skyblue', edgecolor='black')

# Customise the plot
ax.set_title('Top 10 Locations with the Most Earthquakes')
ax.set_xlabel('Place')
ax.set_ylabel('Number of Earthquakes')
plt.xticks(rotation=45, ha='right')

# Adjust layout and display the plot
plt.tight_layout()
```

```python
import random

# Group data by location and count the number of events
location_counts = data['location'].value_counts()

# Focus on the top 5 locations and combine the rest as 'Other'
top_locations = location_counts.head(5)
other_count = location_counts.iloc[5:].sum()
top_locations['Other'] = other_count

# Custom colour palette (hex codes)
custom_colours = ['#9F7257', '#9F2B00', '#47842E', '#E7AD41', '#D3A88F', '#DF7425']

# Create figure and axes
fig, ax = plt.subplots(figsize=(10,8))

# Plot the pie chart
ax.pie(
    top_locations.values,
    labels=top_locations.index,
    autopct='%1.1f%%',  # Show percentages
    startangle=140,     # Start angle for better visualisation
    colors=custom_colours,  # Use custom colours
    textprops={'fontsize':12}         # Adjust font size
)

# Add title and legend
ax.set_title('Proportion of Events by Location', fontsize=18, fontweight='bold')
ax.legend(title='Locations', bbox_to_anchor=(1.05, 1), loc='upper left')

# Optimise layout
plt.tight_layout()
```

```python

# Group by location and date to count events
time_series = data.groupby(['location', 'date']).size().reset_index(name='event_count')

# Focus on the top 5 most frequent locations
top_locations = data['location'].value_counts().head(5).index
filtered_data = time_series[time_series['location'].isin(top_locations)]

# Define line styles and colours
line_styles = ['-', '--', '-.', ':', '-']

# Custom colour palette (hex codes)
custom_colours = ['#9F7257', '#9F2B00', '#47842E', '#E7AD41', '#DF7425']

# Set figure and axes
fig, ax = plt.subplots(figsize=(14, 8))

# Plot each location with a distinct line style and colour
for idx, location in enumerate(top_locations):
    location_data = filtered_data[filtered_data['location'] == location]
    if location_data.empty:
        continue  # Skip if no data for this location
    
    # Plot raw event counts
    ax.plot(location_data['date'], location_data['event_count'], 
             label=location, 
             linestyle=line_styles[idx % len(line_styles)], 
             color=custom_colours[idx % len(custom_colours)])
    
    # Annotate the highest point
    max_point = location_data.loc[location_data['event_count'].idxmax()]
    ax.text(max_point['date'], max_point['event_count'], 
             f"{int(max_point['event_count'])}", 
             fontsize=10, color=custom_colours[idx % len(custom_colours)])

# Add title, labels, and legend
ax.set_title('Event Frequency Over Time by Location', fontsize=18, fontweight='bold')
ax.set_xlabel('Date', fontsize=14)
ax.set_ylabel('Event Count', fontsize=14)
ax.legend(title='Location', fontsize=12, loc='upper left')
ax.grid(True, alpha=0.5)

# Format the x-axis dates
ax.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%d-%m-%Y'))
plt.xticks(rotation=45, fontsize=10)

# Optimise layout
plt.tight_layout()
```
